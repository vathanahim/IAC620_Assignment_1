```python
#grey scale image goes from 0-255 where - is black and 255 is white --> 
#the normalize function just normalize it to a value of up to 1

img_apple = cv.normalize(img_apple.astype('float'), None, 0.0, 1.0, cv.NORM_MINMAX)*255
```

```python
print("statistics", img_apple.min(), img_apple.max(), img_apple.std(), img_apple.mean())
```

```python
img_apple_gray = cv.cvtColor(img_apple, cv.COLOR_BGR2GRAY)
```

```python
(height, width) = img_apple_gray.shape
```

```python

tmp = np.zeros((height, width), np.uint8)
img_mean = img_apple.mean()
for i in range(height):
    for j in range(width):
        if(img_apple[i][j]<img_mean):
            tmp[i][j] = 0
        else:
            tmp[i][j] = 255
```
